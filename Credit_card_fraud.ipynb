```python
# Checking for missing data

total = credit_card_fraud_df.isnull().sum().sort_values(ascending = False)
percent = (credit_card_fraud_df.isnull().sum()/credit_card_fraud_df.isnull().count()*100).sort_values(ascending = False)
pd.concat([total, percent], axis=1, keys=['Total', 'Percent']).transpose()

```

```python
# Generate our categorical variable list
credit_card_cat = credit_card_fraud_df.dtypes[credit_card_fraud_df.dtypes == "object"].index.tolist()

credit_card_fraud_df[credit_card_cat].nunique()
```

```python
# Checking Credit Card transactions by time

class_0 = credit_card_fraud_df.loc[credit_card_fraud_df['Class'] == 0]["Time"]
class_1 = credit_card_fraud_df.loc[credit_card_fraud_df['Class'] == 1]["Time"]

hist_data = [class_0, class_1]
group_labels = ['Not Fraud', 'Fraud']

fig = ff.create_distplot(hist_data, group_labels, show_hist=False, show_rug=False)
fig['layout'].update(title='Credit Card Transactions Time Density Plot', xaxis=dict(title='Time [s]'))
iplot(fig, filename='dist_only')

```

```python
# Creating a new Credit Card Fraud Dateframe with Rename the columns  'Hour', 'Class', 'Min', 'Max'. 'Transactions', 'Sum', 'Mean', 'Median', 'Var'
credit_card_fraud_df['Hour'] = credit_card_fraud_df['Time'].apply(lambda x: np.floor(x / 3600))

tmp = credit_card_fraud_df.groupby(['Hour', 'Class'])['Amount'].aggregate(['min', 'max', 'count', 'sum', 'mean', 'median', 'var']).reset_index()
new_credit_card_fraud_df = pd.DataFrame(tmp)
new_credit_card_fraud_df.columns = ['Hour', 'Class', 'Min', 'Max', 'Transactions', 'Sum', 'Mean', 'Median', 'Var']
new_credit_card_fraud_df.head()
```

```python
# Checking for the total amount of transactions per hour using 'Class'
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18,6))
s = sns.lineplot(ax = ax1, x="Hour", y="Transactions", data=new_credit_card_fraud_df.loc[new_credit_card_fraud_df.Class==0])
s = sns.lineplot(ax = ax2, x="Hour", y="Transactions", data=new_credit_card_fraud_df.loc[new_credit_card_fraud_df.Class==1], color="red")
plt.suptitle("Total Number of Transactions")
plt.show();
```

```python
# Let's plot the fraudulent transactions (amount) against time. 
# The time is shown is seconds from the start of the time period (totaly 48h, over 2 days).

fraud = credit_card_fraud_df.loc[credit_card_fraud_df['Class'] == 1]

trace = go.Scatter(
    x = fraud['Time'],y = fraud['Amount'],
    name="Amount",
     marker=dict(
                color='rgb(238,23,11)',
                line=dict(
                    color='red',
                    width=1),
                opacity=0.5,
            ),
    text= fraud['Amount'],
    mode = "markers"
)
data = [trace]
layout = dict(title = 'Amount of fraudulent transactions',
          xaxis = dict(title = 'Time [s]', showticklabels=True), 
          yaxis = dict(title = 'Amount'),
          hovermode='closest'
         )
fig = dict(data=data, layout=layout)
iplot(fig, filename='fraud-amount')

```
